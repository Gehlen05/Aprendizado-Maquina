```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
df = pd.read_csv('Fish.xls')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.heatmap(df.corr(), annot = True);
```

```python
sns.pairplot(df)
plt.show;
```

```python
y = df.Weight.values.reshape(-1,1)
x = df.Width.values.reshape(-1,1)

plt.scatter(x,y)
plt.ylabel("Massa (g)")
plt.xlabel("Comprimento diagonal (cm)")
plt.title("Rela√ß√£o massa por comprimento")
```

```python
polynomial_features= PolynomialFeatures(degree=2)
x_poly = polynomial_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)


plt.scatter(x, y, s=10)
plt.plot(x, y_poly_pred, color='m')
plt.show()
```

RMSE

```python
np.sqrt(mean_squared_error(y,y_poly_pred))
```

R2

```python
r2_score(y,y_poly_pred)
```

VIF

```python
1/(1-r2_score(y,y_poly_pred))
```
